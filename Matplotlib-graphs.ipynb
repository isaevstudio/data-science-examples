# Visualizing & Customizing A Boxplot

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

matplotlib.interactive(True)
plt.ion()
matplotlib.is_interactive()
```

```python
x = np.random.randint(low=0, high=20, size=20)
x.sort()
x
```

```python
plt.boxplot(x);
```

```python
x = np.append(x, 22)

plt.boxplot(x);
```

```python
x = np.append(x, 55)
x = np.append(x, 45)
# outliers are represented by the fliers; dots are called fliers
plt.boxplot(x);
```

```python
plt.boxplot(x, vert=False);
```

```python
# focus on median (more visible)
plt.boxplot(x, vert=False, notch=True);
```

```python
# Hide outliers
plt.boxplot(x, vert=False, notch=True, showfliers=False);
```

```python
# converting to 2D patch object
# This boxplot patch is basically a dictionary that allows us to access the individual components of this 2D patch
bp = plt.boxplot(x, patch_artist=True)
bp;
```

```python
print('Number of boxes:', len(bp['boxes']))
print('Number of whiskers:', len(bp['whiskers']))
print('Number of caps:', len(bp['caps']))
print('Number of medians:', len(bp['medians']))
print('Number of fliers:', len(bp['fliers']))
```

```python
bp = plt.boxplot(x, patch_artist=True)

bp['boxes'][0].set(facecolor='yellow', edgecolor='blue', hatch='\\')
bp['medians'][0].set(color='red')
```

```python
bp = plt.boxplot(x, patch_artist=True)

bp['boxes'][0].set(facecolor='yellow', edgecolor='blue', hatch='\\')

bp['whiskers'][0].set(color='r', linewidth=3) # bottom
bp['whiskers'][1].set(color='purple', linewidth=4) # top

bp['caps'][0].set(color='yellow', linewidth=2) # top
bp['caps'][1].set(color='green', linewidth=2.5) # bottom

bp['medians'][0].set(color='red')
```

```python
bp = plt.boxplot(x, patch_artist=True)

bp['boxes'][0].set(facecolor='yellow', edgecolor='blue', hatch='\\')

bp['whiskers'][0].set(color='r', linewidth=3) # bottom
bp['whiskers'][1].set(color='purple', linewidth=4) # top

bp['caps'][0].set(color='yellow', linewidth=2) # bottom
bp['caps'][1].set(color='green', linewidth=2.5) # top

bp['fliers'][0].set(marker='D', markerfacecolor='yellow', markeredgecolor='red')

bp['medians'][0].set(color='red', linestyle='--', linewidth=2.5)
```

# Using boxplot in real-world dataset

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

matplotlib.interactive(True)
plt.ion()
matplotlib.is_interactive()
```

```python
exam_data = pd.read_csv('exams.csv')
exam_data
```

```python
exam_scores = exam_data[['math score', 'reading score', 'writing score']]
exam_scores
```

```python
exam_scores_array = np.array(exam_scores)
exam_scores_array
```

```python
bp = plt.boxplot(exam_scores_array)
```

```python
bp = plt.boxplot(exam_scores_array, patch_artist=True)
```

```python
print('Number of boxes:', len(bp['boxes']))
print('Number of whiskers:', len(bp['whiskers']))
print('Number of caps:', len(bp['caps']))
print('Number of medians:', len(bp['medians']))
print('Number of fliers:', len(bp['fliers']))
```

```python
colors = ['blue', 'grey', 'lawngreen']

bp = plt.boxplot(exam_scores_array, patch_artist=True)

for i in range(len(bp['boxes'])):
    
    bp['boxes'][i].set(facecolor=colors[i])
    
    bp['caps'][2*i + 1].set(color=colors[i]) # whiskers 1,2,3,4,5,6 -->> we selected 1,3,5 (1,2 color1; 3,4 color2; 5,6 color3)
```

```python
colors = ['blue', 'grey', 'lawngreen']

bp = plt.boxplot(exam_scores_array, patch_artist=True)

for i in range(len(bp['boxes'])):
    print('boxplot:',i)
    
    bp['boxes'][i].set(facecolor=colors[i])
    
    bp['caps'][2*i + 1].set(color=colors[i])
    
plt.xticks([1,2,3], ['Math', 'Reading', 'Writing']);
```

# Violin PLots

```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

matplotlib.interactive(True)
plt.ion()
matplotlib.is_interactive()
```

```python
exam_data = pd.read_csv('exams.csv')
exam_data.head()
```

```python
exam_scores = exam_data[['math score', 'reading score', 'writing score']]
exam_scores.head()
```

```python
exam_scores_array = np.array(exam_scores)
exam_scores_array
```

```python
vp = plt.violinplot(exam_scores_array)
```

```python
vp = plt.violinplot(exam_scores_array, showmedians=True)

plt.xticks([1,2,3], ['Math', 'Reading', 'Writing']);
```

```python
vp = plt.violinplot(exam_scores_array, showmedians=True, vert=False)

plt.yticks([1,2,3], ['Math', 'Reading', 'Writing']);
```

```python
colors = ['blue', 'grey', 'lawngreen']

vp = plt.violinplot(exam_scores_array, showmedians=True, vert=False)

plt.yticks([1,2,3], ['Math', 'Reading', 'Writing'])

for i in range(len(vp['bodies'])):
    vp['bodies'][i].set(facecolor=colors[i]);
```

```python
colors = ['blue', 'grey', 'lawngreen']

vp = plt.violinplot(exam_scores_array, showmedians=True, vert=False)

plt.yticks([1,2,3], ['Math', 'Reading', 'Writing'])

for i in range(len(vp['bodies'])):
    
    vp['bodies'][i].set(facecolor=colors[i])
    
vp['cmaxes'].set(color='green')
vp['cmins'].set(color='red')
vp['cbars'].set(linestyle=':')
vp['cmedians'].set(linewidth=6);
```

```python
colors = ['blue', 'grey', 'lawngreen']

vp = plt.violinplot(exam_scores_array, showmedians=True, vert=False)

plt.yticks([1,2,3], ['Math', 'Reading', 'Writing'])

for i in range(len(vp['bodies'])):
    
    vp['bodies'][i].set(facecolor=colors[i])
    
vp['cmaxes'].set(color='green')
vp['cmins'].set(color='red')
vp['cbars'].set(linestyle=':')
vp['cmedians'].set(linewidth=6)

plt.legend(handles = [vp['bodies'][0], vp['bodies'][1]],
           labels=['Math', 'Reading'],
           loc='upper left');
```

# Histogram

```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

matplotlib.interactive(True)
plt.ion()
matplotlib.is_interactive()
```

```python
np_data = pd.read_csv('national_parks.csv')
np_data.head()
```

```python
# Tallest = there are 16 years where the number of visitors are between 4.2 and 4.7 millions
# about 16 years 4.2-4.7 millions of visitors have visited Grand Canyon
plt.hist(np_data['GrandCanyon'], facecolor='orange', edgecolor='green', bins=10);
```

```python
n, bins, patches = plt.hist(np_data['GrandCanyon'], facecolor='orange', edgecolor='green', bins=10);

print('n:', n)
# first bin - the frequency (number) of data points are 5
print('bins:', bins)
print('patches:', patches)
```

```python
n, bins, patches = plt.hist(np_data['GrandCanyon'], facecolor='orange', edgecolor='green', bins=10, density=True);

print('n:', n)
print('bins:', bins)
print('patches:', patches)
```

```python
# Here is the calculation
# E.g. In our histogram the largest bin contains 16 of the 57 data points (28%). The interval for the histogram is 
# about 5M (6.25M - 1.25M) and there are 10 bins, so about 500,000 for each bin.

# When converted to probability densities, the value of the largest bin is 5.6e-7. The probability for that
# bin is 500000*5.6e-7 = 0.28
```

```python
plt.hist(np_data['GrandCanyon'], facecolor='orange', edgecolor='green', bins=10, cumulative=True);
```

```python
# Overlapping histograms
plt.hist(np_data['BryceCanyon'], facecolor='maroon', edgecolor='black', bins=8)

plt.hist(np_data['GrandCanyon'], facecolor='orange', edgecolor='green', bins=10);
```

```python
# Overlapping histograms
plt.hist(np_data['BryceCanyon'], facecolor='maroon', edgecolor='black', bins=8)

plt.hist(np_data['GrandCanyon'], facecolor='orange', edgecolor='green', bins=10, alpha=0.4);
```

# Pie Charts

```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

matplotlib.interactive(True)
plt.ion()
matplotlib.is_interactive()
```

```python
data = pd.read_csv('sector_weighting.csv')
data.head()
```

```python
plt.pie(data['Percentage'], labels=data['Sector'])
```

```python
# if the circle is a little bit skewed use plt.axis('equal') to deal it
# chart starts from right side (3 o'clock) and goes anti-clockwise
plt.pie(data['Percentage'], labels=data['Sector'])
plt.axis('equal')
```

```python
colors = ['lawngreen', 'orange', 'teal', 'yellow', 'violet']
```

```python
# adding numbers on chart
plt.pie(data['Percentage'], labels=data['Sector'], colors=colors, autopct='%.1f')
plt.axis('equal');
```

```python
# start from zero and go clockwise
plt.pie(data['Percentage'], labels=data['Sector'], colors=colors, autopct='%.1f', startangle=90, counterclock=False)
plt.axis('equal');
```

```python
# Seperate the pie parts
# a fraction of the radius with which to offset each wedge
explode = (0, 0.1, 0, 0.3, 0) # 1. 0% explode; 2. 10% explode from the center...

plt.pie(data['Percentage'],
        labels=data['Sector'],
        colors=colors,
        autopct='%.1f',
        explode=explode)

plt.axis('equal');
```

```python
wedges, texts, autotexts = plt.pie(data['Percentage'],
                                 labels=data['Sector'],
                                 colors=colors,
                                 autopct='%.1f',
                                 explode=explode)

plt.axis('equal');

# print('Wedges:', wedges)
print('Text:', texts)
print('Autotexts:', autotexts);
```

```python
wedges, texts, autotexts = plt.pie(data['Percentage'],
                                 labels=data['Sector'],
                                 colors=colors,
                                 autopct='%.1f',
                                 explode=explode)

plt.axis('equal')

wedges[1].set(edgecolor='purple', linewidth=2)
texts[1].set(family='monospace', size=20)
autotexts[1].set(weight='bold', size=15);
```

# Plotting Autocorrelations in Time Series Data

__Correlation__ is the relationship between two variables <br>
__Autocorrelation__ auto(self) is the relationship of any variable with itself (Measure the relationship between a variable's `current` and `past` value)

```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

matplotlib.interactive(True)
plt.ion()
matplotlib.is_interactive()
```

```python
grand_canyon_data = pd.read_csv('grand_canyon_visits.csv')
grand_canyon_data.head()
```

```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

matplotlib.interactive(True)
plt.ion()
matplotlib.is_interactive()

data = pd.read_csv('sector_weighting.csv')
data.head()
```

```python
grand_canyon_data['NumVisits'].describe()
```

```python
# large numbers in autocorrelation might result an overflow during mathematical operations;
# To avoid this we divide the number of visits by 1000;
grand_canyon_data['NumVisits'] = grand_canyon_data['NumVisits'] / 1000
grand_canyon_data['NumVisits'].describe()
```

```python
# We are going to use 20 logs -->> range between -20 & 20 
plt.figure(figsize=(16, 8))

plt.acorr(grand_canyon_data['NumVisits'], maxlags=20);  # we take the data and lag it 20 times; we end up with time series
# the autocorrelation is highest when the lag is 12 (-12&12) it shows the seasonality in park visits
```

```python
plt.figure(figsize=(16, 8))

lags, c, vlines, hline = plt.acorr(grand_canyon_data['NumVisits'], maxlags=20);

print('Lags:', lags, '\n')
print('C:', c, '\n') # contains the correlation for each of the lag values
print('Vlines:', vlines, '\n')
print('Hline:', hline, '\n')
```

# Stacked Plots & Stem Plots

```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

matplotlib.interactive(True)
plt.ion()
matplotlib.is_interactive()
```

```python
np_data = pd.read_csv('national_parks.csv')
np_data.head()
```

```python
x = np_data['Year']
```

```python
y = np.vstack([np_data['Badlands'],
               np_data['GrandCanyon'],
               np_data['BryceCanyon']])

y
```

```python
# Stack plot means the plots that are one on top of the other; done by np.vstack()
labels = ['Badlands','GrandCanyon','BryceCanyon']
```

# 

#### Stacked plots

```python
plt.stackplot(x, y, labels=labels)

plt.legend(loc='upper left');
```

```python
colors = ['lawngreen', 'orange', 'teal']
```

```python
plt.stackplot(x, y, labels=labels, colors=colors, edgecolor='black')

plt.legend(loc='upper left');
```

# 

#### Stem plots

It is a good __data visualization__ technique to view __fluctuating data__ # fluctuate - kolebatsya

```python
# will subtract the number of visitors, current number of visitors from the previous year number of visitors for us
np_data[['Badlands', 'GrandCanyon', 'BryceCanyon']] = np_data[['Badlands', 'GrandCanyon', 'BryceCanyon']].diff()

np_data
# negative number means the visitor number dropped significantly 
```

```python
# stem plots are the best plots for visualizing the flactuating data; Cause it can flactuate in pos. and neg. numbers
plt.figure(figsize=(10,6))

plt.stem(np_data['Year'], np_data['Badlands'])

plt.title('Change in number of visitors');
```

```python
plt.figure(figsize=(10,6))

plt.stem(np_data['Year'],
         np_data['Badlands'],
         markerfmt='r_', # first color second style
         linefmt='g--',
         basefmt=':') 

plt.title('Change in number of visitors');

# stems going up are possitive deltas; Going down are negavite deltas
```
